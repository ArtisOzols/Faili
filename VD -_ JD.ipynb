```python
from pathlib import Path

def getFilter(path="./", prefix="", match="", ignore="", suffix=""):
    p = Path(path)
    if ignore:
        return [f for f in p.iterdir() if f.is_file() and f.name.startswith(prefix) and match in f.name and not ignore in f.name and f.name.endswith(suffix)]
    return [f for f in p.iterdir() if f.is_file() and f.name.startswith(prefix) and match in f.name and f.name.endswith(suffix)]
```

```python
prefixes = ["iſ", "is", "aiſ", "ais", "uſ", "us", "at", "ap", "beſ", "bes", "ie", "neſ", "nes", "no", "pa", "pahr", "pie", "sa", "jah", "Iſ", "Is", "Aiſ", "Ais", "Uſ", "Us", "At", "Ap", "Beſ", "Bes", "Ie", "Neſ", "Nes", "No", "Pa", "Pahr", "Pie", "Sa", "Jah"]
vowels = ["a", "e", "i", "o", "u"]
def add_letters(key, val, d, letters=[""], before_key=0):
    if before_key:
        for i in letters:
            d[i+key] = i+val
    else:
        for i in letters:
            d[key+i] = val+i
    return d

def starts_ends_with(key, val, text, starts_ends=0):
    text = " " + text + " "
    s_list = [" ", "(", "[", '"', "/", "\n", "\t"] if starts_ends == 1 or starts_ends == 0 else [""]
    e_list = [" ", ",", ".", "!", "?", ":", ";", ")", "]", "/", "\n", "\t"] if starts_ends == 2 or starts_ends == 0 else [""]

    for s in s_list: 
        for e in e_list:
            text = text.replace(s + key + e, s + val + e)
            if not key.istitle():
                text = text.replace(s + key.capitalize() + e, s + val.capitalize() + e)
    return text[1:-1]

symb_list = ["𓀠", "𓀡", "𓀢", "𓀣", "𓀤", "𓀥", "𓀀", "𓀁", "𓀂", "𓀃", "𓀐", "𓀑", "𓀒", "𓀓", "𓀔", "𓀕", "𓀉", "𓀊", "𓀋", "𓀞", "𓀟", "𓁍", "𓁎", "𓁏", "𓁨", "𓁋", "𓇀", "𓇁", "𓇂", "𓇃", "𓇄", "𓇅", "𓇆", "𓇇", "𓇈", "𓇉", "𓇊", "𓇋", "𓇌", "𓇍", "𓇎", "𓇏", "𓆐", "𓆑", "𓆒", "𓆓", "𓆔", "𓆕", "𓆖", "𓆗", "𓆘", "𓆙", "𓆚", "𓆛", "𓆜", "𓆝", "𓆞", "𓆟", "𓋀", "𓋁", "𓋂", "𓋃", "𓋄", "𓋅", "𓋆", "𓋇", "𓋈", "𓋉", "𓋊", "𓋋", "𓋌", "𓋍", "𓋎", "𓋏", "𓁄", "𓁁"]
def exceptions(e_list, dict, decode):
    i = 0
    if not decode:
        for key in e_list:
            dict[key] = symb_list[i]
            i +=1
    else:
        for key in e_list:
            dict[symb_list[i]] = key
            i +=1
    
    return dict
        
# https://www.fileformat.info/info/unicode/block/egyptian_hieroglyphs/utf8test.htm
```
